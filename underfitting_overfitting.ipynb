```python
import numpy as np
import pandas as pd
import sklearn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
from sklearn.datasets import make_moons
import plotly.express as px
np.random.seed(42)
```

```python
raw_data = make_moons(n_samples = 2000, noise = 0.25, random_state = 42)

data = raw_data[0]
target = raw_data[1]

print(data[:5])
print(target[:5])
```

```python
df = pd.DataFrame(data = np.c_[data, target], columns = ['x1', 'x2', 'target'])
df.head()
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(data[:, 0], data[:, 1], c = target, cmap = 'viridis')
plt.title('Dane do modelu klasyfikacji')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.3)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(X_train[:, 0], X_train[:, 1], c = y_train, cmap = 'RdYlBu', label = 'training_set')
plt.scatter(X_test[:, 0], X_test[:, 1], c = y_test, cmap = 'RdYlBu', marker = 'x', alpha = 0.5, label = 'test_set')
plt.title('Zbir treningowy i testowy')
plt.legend()
plt.show()
```

Niedouczenie - Regresja logistyczna

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(10,8))
plot_decision_regions(X_train, y_train, classifier)
plt.title(f'Zbir treningowy o dokadnoci: {classifier.score(X_train, y_train) :.4f}')
```

```python
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(10,8))
plot_decision_regions(X_test, y_test, classifier)
plt.title(f'Zbir testowy o dokadnoci: {classifier.score(X_test, y_test) :.4f}')
```

niedouczenie - drzewo decyzyjne

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(max_depth = 1)
classifier.fit(X_train, y_train)

plt.figure(figsize=(10,8))
plot_decision_regions(X_train, y_train, classifier)
plt.title(f'Zbir treningowy o dokadnoci: {classifier.score(X_train, y_train) :.4f}')
```

```python
plt.figure(figsize=(10,8))
plot_decision_regions(X_test, y_test, classifier)
plt.title(f'Zbir testowy o dokadnoci: {classifier.score(X_test, y_test) :.4f}')
```

Przykad przuczenia

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)

plt.figure(figsize=(10,8))
plot_decision_regions(X_train, y_train, classifier)
plt.title(f'Zbir treningowy o dokadnoci: {classifier.score(X_train, y_train) :.4f}')
```

```python
plt.figure(figsize=(10,8))
plot_decision_regions(X_test, y_test, classifier)
plt.title(f'Zbir testowy o dokadnoci: {classifier.score(X_test, y_test) :.4f}')
```
