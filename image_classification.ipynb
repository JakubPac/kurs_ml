```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from tensorflow.keras.datasets.fashion_mnist import load_data
```

```python
np.set_printoptions(precision = 12, suppress = True, linewidth = 150)
```

```python
sns.set()
```

```python
(X_train, y_train), (X_test, y_test) = load_data()
```

```python
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
X_train[0]
```

```python
y_train[:10]
```

```python
plt.imshow(X_train[0], cmap = 'Grays')
plt.axis('off')
```

```python
class_names = ['T-shirt', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker','Bag', 'Ankle boot']

plt.figure(figsize = (14,8))
for i in range(1, 11):
  plt.subplot(1, 10, i)
  plt.axis('off')
  plt.imshow(X_train[i-1], cmap = 'gray_r')
  plt.title(class_names[y_train[i-1]], color = 'black', fontsize = 16)
plt.show()
```

Przygotowanie danych

```python
X_train = X_train / 255.
X_test = X_test / 255.
```

```python
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
```

```python
X_train = X_train.reshape(60000, 28 * 28)
X_test = X_test.reshape(10000, 28 * 28)

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
```

Budowa modelu

```python
from sklearn.svm import SVC

classifier = SVC()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

Ocena modelu

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import plotly.figure_factory as ff

def plot_confusion_matrix(cm):
  cm = cm[::-1]
  cm = pd.DataFrame(cm, columns = class_names, index = class_names[::-1])

  fig = ff.create_annotated_heatmap(z = cm.values, x = list(cm.columns), y = list(cm.index),
                                    colorscale = 'ice', showscale = True, reversescale = True)
  fig.update_layout(width = 700, height = 500, title = 'Macierz konfuzji', font_size = 16)
  fig.show()

plot_confusion_matrix(cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred, target_names = class_names))
```

bldnie sklasyfikowane obrazy

```python
results = pd.DataFrame({'y_true': y_test, 'y_pred':y_pred})
results.head(10)
```

```python
errors = results[results['y_true'] != results['y_pred']]
errors_indx = list(errors.index)
errors_indx[:10]
```

```python
results.loc[errors_indx, :]
```

```python
plt.figure(figsize = (20,10))
for idx, error_idx in enumerate(errors_indx[:15]):
  image = X_test[error_idx].reshape(28,28)
  plt.subplot(3, 5, idx + 1)
  plt.axis('off')
  plt.imshow(image, cmap = 'Greys')
  plt.title(f'True: {class_names[results.loc[error_idx, "y_true"]]}. Prediction: {class_names[results.loc[error_idx, "y_pred"]]}')
```
